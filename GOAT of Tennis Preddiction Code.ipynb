```python
players = ['Rafael Nadal',
 'Novak Djokovic',
 'Roger Federer',
 'Andre Agassi',
 'Pete Sampras',
 'Stan Wawrinka',
 'Andy Murray',
 'Gustavo Kuerten',
 'Marat Safin',
 'Lleyton Hewitt']


df['year']=pd.to_datetime(df['tourney_date']).dt.year

df_top_players = df[(df['winner_name'].isin(players)) | (df['loser_name'].isin(players))]
df_top_players = df_top_players[df_top_players['tourney_level'] == 'G']

df_top_players
```

#### Here we have extracted the features that we want to consider for our model cummulated for each year for each of the Top 10 player:

```python
yearly_stats = pd.DataFrame()

for year in df_top_players['year'].unique():
    year_data = df_top_players[df_top_players['year'] == year]

    wins = year_data['winner_name'].value_counts().rename("wins").reset_index()
    wins.columns = ['player_name', 'wins']

    losses = year_data['loser_name'].value_counts().rename("loss").reset_index()
    losses.columns = ['player_name', 'loss']

    yearly_win_loss = pd.merge(wins, losses, on='player_name', how='outer').fillna(0)
    yearly_win_loss['wins'] = yearly_win_loss['wins'].astype(int)
    yearly_win_loss['loss'] = yearly_win_loss['loss'].astype(int)

    yearly_win_loss['matches_played'] = yearly_win_loss['wins'] + yearly_win_loss['loss']
    yearly_win_loss['win_percentage'] = yearly_win_loss.apply(
        lambda row: row['wins'] / row['matches_played'] if row['matches_played'] > 0 else pd.NA,
        axis=1
    )
    yearly_win_loss['win/loss ratio'] = yearly_win_loss.apply(
        lambda row: row['wins'] / row['loss'] if row['loss'] > 0 else pd.NA,
        axis=1
    )
    yearly_win_loss['year'] = year

    yearly_win_loss = yearly_win_loss[yearly_win_loss['player_name'].isin(players)]

    win_stats = year_data[year_data['winner_name'].isin(players)].groupby('winner_name').agg(
        w_ace_sum=('w_ace', 'sum'),
        w_df_sum=('w_df', 'sum'),
        w_svpt_sum=('w_svpt', 'sum'),
        w_1stIn_sum=('w_1stIn', 'sum'),
        w_1stWon_sum=('w_1stWon', 'sum'),
        w_2ndWon_sum=('w_2ndWon', 'sum'),
        w_SvGms_sum=('w_SvGms', 'sum'),
        w_bpSaved_sum=('w_bpSaved', 'sum'),
        w_bpFaced_sum=('w_bpFaced', 'sum')
    ).reset_index().rename(columns={'winner_name': 'player_name'})

    loss_stats = year_data[year_data['loser_name'].isin(players)].groupby('loser_name').agg(
        l_ace_sum=('l_ace', 'sum'),
        l_df_sum=('l_df', 'sum'),
        l_svpt_sum=('l_svpt', 'sum'),
        l_1stIn_sum=('l_1stIn', 'sum'),
        l_1stWon_sum=('l_1stWon', 'sum'),
        l_2ndWon_sum=('l_2ndWon', 'sum'),
        l_SvGms_sum=('l_SvGms', 'sum'),
        l_bpSaved_sum=('l_bpSaved', 'sum'),
        l_bpFaced_sum=('l_bpFaced', 'sum')
    ).reset_index().rename(columns={'loser_name': 'player_name'})

    player_stats = pd.merge(win_stats, loss_stats, on='player_name', how='outer').fillna(0)

    player_stats['ace_sum'] = player_stats['w_ace_sum'] + player_stats['l_ace_sum']
    player_stats['df_sum'] = player_stats['w_df_sum'] + player_stats['l_df_sum']
    player_stats['1stIn_sum'] = player_stats['w_1stIn_sum'] + player_stats['l_1stIn_sum']
    player_stats['1stWon_sum'] = player_stats['w_1stWon_sum'] + player_stats['l_1stWon_sum']
    player_stats['2ndWon_sum'] = player_stats['w_2ndWon_sum'] + player_stats['l_2ndWon_sum']
    player_stats['SvGms_sum'] = player_stats['w_SvGms_sum'] + player_stats['l_SvGms_sum']
    player_stats['bpSaved_sum'] = player_stats['w_bpSaved_sum'] + player_stats['l_bpSaved_sum']
    player_stats['bpFaced_sum'] = player_stats['w_bpFaced_sum'] + player_stats['l_bpFaced_sum']
    player_stats['svpt_sum'] = player_stats['w_svpt_sum'] + player_stats['l_svpt_sum']

    yearly_player_stats = pd.merge(yearly_win_loss, player_stats, on='player_name', how='left')

    yearly_stats = pd.concat([yearly_stats, yearly_player_stats], ignore_index=True)

yearly_stats['A%'] = (yearly_stats['ace_sum'] / yearly_stats['svpt_sum']) * 100
yearly_stats['DF%'] = (yearly_stats['df_sum'] / yearly_stats['svpt_sum']) * 100
yearly_stats['1st%'] = (yearly_stats['1stIn_sum'] / yearly_stats['svpt_sum']) * 100
yearly_stats['1stWon%'] = (yearly_stats['1stWon_sum'] / yearly_stats['1stIn_sum']) * 100
yearly_stats['2ndWon%'] = (yearly_stats['2ndWon_sum'] / (yearly_stats['svpt_sum'] - yearly_stats['1stIn_sum'])) * 100
yearly_stats['BP Saved %'] = (yearly_stats['bpSaved_sum'] / yearly_stats['bpFaced_sum']) * 100

yearly_stats = yearly_stats[['year', 'player_name', 'wins', 'loss', 'win_percentage', 'win/loss ratio',
                             'matches_played', 'ace_sum', 'A%', 'df_sum', 'DF%', '1stIn_sum', '1st%',
                             '1stWon_sum', '1stWon%', '2ndWon_sum', '2ndWon%', 'SvGms_sum', 'bpSaved_sum',
                             'bpFaced_sum', 'BP Saved %', 'svpt_sum']]

yearly_stats
```

```python
df_top_players['sets_won'] = 0
df_top_players['sets_lost'] = 0

for idx, row in df_top_players.iterrows():
    sets = str(row['score']).split()
    winner_sets_won = 0
    loser_sets_won = 0

    for set_score in sets:
        if '-' in set_score:
            winner_games, loser_games = set_score.split('-')
            winner_games = int(winner_games.split('(')[0])
            loser_games = int(loser_games.split('(')[0])

            if winner_games > loser_games:
                winner_sets_won += 1
            elif loser_games > winner_games:
                loser_sets_won += 1

    df_top_players.at[idx, 'sets_won'] = winner_sets_won
    df_top_players.at[idx, 'sets_lost'] = loser_sets_won

yearly_sets = df_top_players.groupby(['year', 'winner_name']).agg(
    sets_won_total=('sets_won', 'sum'),
    sets_lost_total=('sets_lost', 'sum')
).reset_index()

yearly_sets.rename(columns={'winner_name': 'player_name'}, inplace=True)

yearly_stats = pd.merge(yearly_stats, yearly_sets, on=['year', 'player_name'], how='left')

yearly_stats[['sets_won_total', 'sets_lost_total']] = yearly_stats[['sets_won_total', 'sets_lost_total']].fillna(0)
```

#### Normalizing the data using MinMax Scaler

```python
from sklearn.preprocessing import MinMaxScaler

feature_columns = [
    'wins', 'loss', 'win_percentage', 'win/loss ratio', 'matches_played', 'ace_sum', 'A%',
    'df_sum', 'DF%', '1stIn_sum', '1st%', '1stWon_sum', '1stWon%', '2ndWon_sum', '2ndWon%',
    'SvGms_sum', 'bpSaved_sum', 'bpFaced_sum', 'BP Saved %', 'sets_won_total',
    'sets_lost_total', 'total_sets_played', 'sets_win_percentage','svpt_sum', 'Initial Rank Inverted',	'Final Rank Inverted'
]

scaler = MinMaxScaler()
normalized_features = scaler.fit_transform(new_final_df[feature_columns])

normalized_df = new_final_df.copy()
normalized_df[feature_columns] = normalized_features

normalized_df.head()
```

#### Using Linear Regression Model

```python
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

features = [
    'wins', 'loss', 'win_percentage', 'win/loss ratio', 'matches_played',
    'ace_sum', 'A%', 'df_sum', 'DF%', '1stIn_sum', '1st%', '1stWon_sum',
    '1stWon%', '2ndWon_sum', '2ndWon%', 'SvGms_sum', 'bpSaved_sum', 'bpFaced_sum',
    'BP Saved %', 'sets_won_total', 'sets_lost_total', 'total_sets_played',
    'sets_win_percentage', 'Initial Rank Inverted'
]

X = normalized_df[features]
y = normalized_df['Final Rank Inverted']

model = LinearRegression()
model.fit(X, y)

y_pred = model.predict(X)
mse = mean_squared_error(y, y_pred)
r2 = r2_score(y, y_pred)

print("Mean Squared Error:", mse)
print("R^2 Score:", r2)
print("Coefficients:", model.coef_)
print("Intercept:", model.intercept_)
```

```python
career_stats = new_final_df.groupby('player_name').agg(
    wins=('wins', 'sum'),
    loss=('loss', 'sum'),
    matches_played=('matches_played', 'sum'),
    ace_sum=('ace_sum', 'sum'),
    df_sum=('df_sum', 'sum'),
    **{'1stIn_sum': ('1stIn_sum', 'sum')},
    **{'1stWon_sum': ('1stWon_sum', 'sum')},
    **{'2ndWon_sum': ('2ndWon_sum', 'sum')},
    SvGms_sum=('SvGms_sum', 'sum'),
    bpSaved_sum=('bpSaved_sum', 'sum'),
    bpFaced_sum=('bpFaced_sum', 'sum'),
    sets_won_total=('sets_won_total', 'sum'),
    sets_lost_total=('sets_lost_total', 'sum'),
    total_sets_played=('total_sets_played', 'sum'),
    svpt_sum=('svpt_sum', 'sum'),
    Initial_Rank=('Initial Rank', 'mean'),
    Final_Rank=('Final Rank', 'mean'),
).reset_index()

career_stats['win_percentage'] = career_stats['wins'] / career_stats['matches_played']
career_stats['win/loss ratio'] = career_stats['wins'] / career_stats['loss']

career_stats['A%'] = (career_stats['ace_sum'] / career_stats['svpt_sum']) * 100
career_stats['DF%'] = (career_stats['df_sum'] / career_stats['svpt_sum']) * 100
career_stats['1st%'] = (career_stats['1stIn_sum'] / career_stats['svpt_sum']) * 100
career_stats['1stWon%'] = (career_stats['1stWon_sum'] / career_stats['1stIn_sum']) * 100
career_stats['2ndWon%'] = (career_stats['2ndWon_sum'] / (career_stats['svpt_sum'] - career_stats['1stIn_sum'])) * 100
career_stats['BP Saved %'] = (career_stats['bpSaved_sum'] / career_stats['bpFaced_sum']) * 100
career_stats['sets_win_percentage'] = career_stats['sets_won_total'] / career_stats['total_sets_played'] * 100
career_stats['sets_win_loss_ratio'] = career_stats['sets_won_total'] / career_stats['sets_lost_total'] * 100
career_stats['Initial_rank_inverted'] = 1 / (career_stats['Initial_Rank'] + 1e-6)
career_stats['Final_rank_inverted'] = 1 / (career_stats['Final_Rank'] + 1e-6)

new_column_order = ['player_name', 'wins', 'loss', 'win_percentage',
       'win/loss ratio', 'matches_played', 'ace_sum', 'A%', 'df_sum', 'DF%',
       '1stIn_sum', '1st%', '1stWon_sum', '1stWon%', '2ndWon_sum', '2ndWon%',
       'SvGms_sum', 'bpSaved_sum', 'bpFaced_sum', 'BP Saved %',
       'sets_won_total', 'sets_lost_total', 'total_sets_played',
       'sets_win_percentage', 'svpt_sum', 'Initial_Rank', 'Final_Rank',
       'Initial_rank_inverted', 'Final_rank_inverted']
career_stats = career_stats[new_column_order]
career_stats
```
